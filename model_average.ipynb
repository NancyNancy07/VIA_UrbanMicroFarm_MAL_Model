This model represents the a baseline, using the average, as a fallback since our datas set is incompatible (the prediction was unreliable). The average would be more accurate than what our attempts using data set managed to produce. 

Calculation: The arithmetic mean of the target column was calculated, excluding NaN values.
Excluded: Rows with missing water data were skipped.
Logic: The model ignores all input features (Temperature, Humidity, Light, etc.) and outputs the pre-calculated mean for every request.

```python
import pandas as pd

import numpy as np
import pickle
```

```python
df = pd.read_csv("Indoor_Plant_Health_and_Growth_Factors.csv")

print("Available columns:")
print(df.columns.tolist())

print("\nFirst 5 rows:")
print(df.head())
```

```python
target_column = "Watering_Amount_ml"

average_value = df[target_column].mean()
data_points = df[target_column].count()
nan_count = df[target_column].isna().sum()

print(f"\nTraining Data Summary:")
print(f"  - Total records: {len(df)}")
print(f"  - Valid data points: {data_points}")
print(f"  - NaN values skipped: {nan_count}")
print(f"  - Calculated average: {average_value:.2f} ml")
```

```python
class AverageWaterModel:
    """
    baseline model that predicts the average water amount.
    Ignores all input features and returns the pre-calculated mean.
    """
    def __init__(self, avg_value):
        self.avg_value = avg_value
        self.model_type = "Average_Baseline"
    
    def fit(self, X, y=None):
        print("Model already fitted with pre-calculated average.")
        return self
    
    def predict(self, X):
        """
        Predict water amount for input data.
        Returns the same average value for all inputs.
        """
        n_samples = len(X) if hasattr(X, '__len__') else 1
        return np.full(n_samples, self.avg_value)
    
    def get_info(self):
        """Return model metadata."""
        return {
            "model_type": self.model_type,
            "predicted_value": self.avg_value,
            "unit": "ml"
        }
```

```python
model = AverageWaterModel(average_value)
```

```python
model_filename = "average_water_model.pkl"
with open(model_filename, 'wb') as f:
    pickle.dump(model, f)

print(f"\n Model saved to: {model_filename}")
```

validation + tetsing

```python
test_input_single = pd.DataFrame({'Temperature': [22]})
pred_single = model.predict(test_input_single)
print(f"\nTest 1 - Single Input:")
print(f"  Input: Temperature=22°C")
print(f"  Prediction: {pred_single[0]:.2f} ml")
print(f"  Passed" if abs(pred_single[0] - average_value) < 0.01 else "  Failed")

```

```python
test_input_multi = pd.DataFrame({
    'Temperature': [20, 25, 30],
    'Humidity': [40, 50, 60]
})
pred_multi = model.predict(test_input_multi)
print(f"\nTest 2 - Multiple Inputs:")
print(f"  Input: 3 different sensor readings")
print(f"  Predictions: {[f'{p:.2f}' for p in pred_multi]}")
all_same = all(abs(p - average_value) < 0.01 for p in pred_multi)
print(f"  Passed" if all_same else "  Failed")

# Test 3: Model info
info = model.get_info()
print(f"\nTest 3 - Model Metadata:")
print(f"  Model Type: {info['model_type']}")
print(f"  Predicted Value: {info['predicted_value']:.2f} {info['unit']}")
print(f" Passed")
```
